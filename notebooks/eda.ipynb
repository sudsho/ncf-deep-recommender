# MovieLens EDA

Quick look at MovieLens-100K before training. Things we want to know:
- How sparse is the user-item interaction matrix?
- Long-tail behaviour: how popular are the popular items, how active are the active users?
- Rating distribution.

Run after `bash scripts/download_data.sh`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

PATH = '../data/ml-100k/u.data'
df = pd.read_csv(PATH, sep='\t', header=None, names=['user', 'item', 'rating', 'ts'])
print(df.shape, df['user'].nunique(), df['item'].nunique())
df.head()
```

## Sparsity

```python
n_users = df['user'].nunique()
n_items = df['item'].nunique()
density = len(df) / (n_users * n_items)
print(f'density = {density:.4f}, sparsity = {1 - density:.4f}')
```

## Popular items (long tail)

```python
item_counts = df.groupby('item').size().sort_values(ascending=False)
plt.figure(figsize=(8, 4))
plt.plot(item_counts.values)
plt.yscale('log')
plt.xlabel('item rank')
plt.ylabel('# interactions (log)')
plt.title('item popularity (long tail)')
plt.show()
print('top 5 items:', item_counts.head().to_dict())
```

## User activity

```python
user_counts = df.groupby('user').size()
user_counts.describe()
```

## Rating distribution

```python
df['rating'].value_counts().sort_index().plot(kind='bar')
plt.title('rating distribution')
plt.show()
```
